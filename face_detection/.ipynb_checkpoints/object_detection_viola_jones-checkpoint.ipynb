## 1. Using pre-trained xml file from OpenCV

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the image
img = cv2.imread("Resources/lenna.png")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# for showing the image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
plt.title('Grayscale Image')
plt.axis("off")
plt.show()
```

```python
# import pre-trained cascade filter file
face_cascade = cv2.CascadeClassifier('Resources/haarcascade_frontalface_default.xml') 
detected_faces = face_cascade.detectMultiScale(gray_img)

# mark all the faces with bounding boxes
for (column, row, width, height) in detected_faces:
    cv2.rectangle(img,(column, row),(column + width, row + height),(0, 255, 0),4)    

# displaying the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

### References
1. P. Viola and M. Jones, "Rapid object detection using a boosted cascade of simple features," Proceedings of the 2001 IEEE Computer Society Conference on Computer Vision and Pattern Recognition. CVPR 2001, Kauai, HI, USA, 2001, pp. I-I, doi: 10.1109/CVPR.2001.990517.

2. Viola Jones Algorithm and Haar Cascade Classifier. https://towardsdatascience.com/viola-jones-algorithm-and-haar-cascade-classifier-ee3bfb19f7d8
