# Object Detection Using Viola-Jones Method
You can detect any object using this method, if you have trained the model. 

I have implemented the following:
1. Face Detection using pre-trained xml file. 
2. Eye Detection using pre-trained xml file.
2. Make a custom XML file for object detection.


## 1. Face detection using pre-trained xml file from opencv

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the image
img = cv2.imread("Resources/lenna.png")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# for showing the image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
plt.title('Grayscale Image')
plt.axis("off")
plt.show()
```

```python
# import pre-trained cascade filter file
face_cascade = cv2.CascadeClassifier('Resources/haarcascade_frontalface_default.xml') 
detected_faces = face_cascade.detectMultiScale(gray_img, scaleFactor = 1.2, minNeighbors = 5)

# mark all the faces with bounding boxes
for (x, y, width, height) in detected_faces:
    cv2.rectangle(img,(x, y),(x + width, y + height),(0, 255, 0),4)    

# displaying the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

## 2. Eye detection using pre-trained haar cascade xml file

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the image
img = cv2.imread("Resources/lenna.png")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# for showing the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis("off")
```

```python
# import pre-trained cascade filter file
eye_cascade = cv2.CascadeClassifier('Resources/haarcascade_eye.xml') 
detected_eyes = eye_cascade.detectMultiScale(gray_img)

# mark all the eyes with bounding boxes
for (x, y, width, height) in detected_eyes:
    cv2.rectangle(img,(x, y),(x + width,y + height),(0, 255, 0),4) 
    
# displaying the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
